```python
#numpy is for scientific computing and matplotlib is for plotting 
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
data = pd.read_csv('/home/rajs11/I_CODE/26/ML/Day2/Housing.csv')
data = data[['price','area']]
data.columns = ['Price','Area']
print(data)
```

```python
#access the data, segregate them into x and y , find m (no. of data)

print({data.shape}) #no. of rows and cols or len()
print(data.shape[0]) 
print(len(data))
m = data.shape[0]
x_train = np.array(data['Area'])
y_train = np.array(data['Price'])
print(x_train[0],y_train[0],m)



```

```python
#plotting 

# marker gives which type of points (x, o ....) , c means color , first is for the y second is for the x
plt.scatter(x_train,y_train,marker='x',c='r')   
plt.title("Housing Prices")
plt.ylabel('Housing Prices (in 1000s of $)')
plt.xlabel('Area is 1000sq ft')
plt.show()
```

```python
#prediction , so as per regression y=mx+c
def prediction_of_price(x,w,b):
    """
    Computes the prediction of a linear model
    Args:
      x (ndarray (m,)): Data, m examples 
      w,b (scalar)    : model parameters  
    Returns
      f_wb (ndarray (m,)): model prediction
    """
    m = x.shape[0]
    pred = np.zeros(m)
    for i in range(m):
        pred[i] = w *x[i]+b
    return pred




#taking random weight and biased value as 100 and 100 initially
w = 1200
b=2300
predicted_price = prediction_of_price(x_train,w,b)


#the predicted price by the model I have now
```

```python
#plotting the price as per the model (reg model)
plt.scatter(x_train,predicted_price,marker='o',c='b',label='our prediction')
plt.scatter(x_train,y_train,marker='x',c='r',label='actual Value')
plt.legend()  #shows the label which all are mentioned above
plt.xlabel('Area in 1000sq ft')
plt.ylabel('Prices in 1000 $')
plt.title('House Prediction')

rj=np.array([1200])
output = prediction_of_price(rj,w,b)
print(output[0])
```
