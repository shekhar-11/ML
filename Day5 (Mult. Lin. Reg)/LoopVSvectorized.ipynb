```python
import numpy as np
import time
```

```python
a = np.random.rand(1000000)
b = np.random.rand(1000000)
# print(a)
```

```python
def loop_version(a, b):
    c = np.zeros(a.shape[0])
    for i in range(a.shape[0]):
        c[i] = a[i] + b[i]
    return c
```

```python
loop_time_start = time.time()
loop_version(a, b)
loop_time_end = time.time()

print(f"Loop version took {loop_time_end - loop_time_start} seconds")
```

```python
dot_time_start = time.time()
dot_version = a + b
dot_time_end = time.time()

print(f"Dot version took {dot_time_end - dot_time_start} seconds")
```
