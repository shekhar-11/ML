```python
import numpy as np
import pandas as pd
%matplotlib inline
```

```python
data = pd.read_csv('/home/rajs11/I_CODE/26/ML/Day2/Housing.csv')
data = data[['price','area']]
x_train = data['area']
y_train = data['price']
```

```python
def prediction_of_price(x,w,b):
    """
    Computes the prediction of a linear model
    Args:
      x (ndarray (m,)): Data, m examples 
      w,b (scalar)    : model parameters  
    Returns
      f_wb (ndarray (m,)): model prediction
    """
    m = x.shape[0]
    pred = np.zeros(m)
    for i in range(m):
        pred[i] = w *x[i]+b
    return pred



def cost_function(x,y,w,b):
    ycap = prediction_of_price(x,w,b)
    m = x.shape[0]
    print(m)
    cf = 0
    for i in range(m):
        cf += (ycap[i]-y[i])**2
    cf = cf /(2*m)

    return cf


print(cost_function(x_train,y_train,1200,2300))
```
